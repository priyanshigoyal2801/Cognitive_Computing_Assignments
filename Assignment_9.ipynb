```python
#Q1
import string

paragraph = """Technology constantly reshapes the world around us. From smartphones to artificial intelligence, it changes how we live, work, and interact. I find it fascinating how rapidly tech evolves and opens up new possibilities every day. Innovations like virtual reality and machine learning are redefining industries. As a tech enthusiast, I love exploring new tools and platforms. It's amazing how technology can solve problems and connect people globally."""

#(1)
clean_text = paragraph.lower().translate(str.maketrans('', '', string.punctuation))
print(clean_text)

#(2)
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize

paragraph = """Technology constantly reshapes the world around us. From smartphones to artificial intelligence, it changes how we live, work, and interact. I find it fascinating how rapidly tech evolves and opens up new possibilities every day. Innovations like virtual reality and machine learning are redefining industries. As a tech enthusiast, I love exploring new tools and platforms. It's amazing how technology can solve problems and connect people globally."""

word_tokens = word_tokenize(paragraph)
sentence_tokens = sent_tokenize(paragraph)

print('\n')
print("Word Tokens:", word_tokens)
print("Sentence Tokens:", sentence_tokens)
print('\n')

nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [word for word in word_tokens if word not in stop_words]

print("Filtered Words (No Stopwords):", filtered_words)
print('\n')


from nltk import FreqDist

fdist = FreqDist(filtered_words)
fdist.plot(10, title="Top 10 Word Frequencies (Excluding Stopwords)")
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

nltk.download('wordnet')

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_porter = [porter.stem(word) for word in filtered_words]
stemmed_lancaster = [lancaster.stem(word) for word in filtered_words]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Porter Stemmer:", stemmed_porter)
print("Lancaster Stemmer:", stemmed_lancaster)
print("Lemmatization:", lemmatized)
```

```python
import re

text = """Technology constantly reshapes the world around us. From smartphones to artificial intelligence, it changes how we live, work, and interact. I find it fascinating how rapidly tech evolves and opens up new possibilities every day. Innovations like virtual reality and machine learning are redefining industries. As a tech enthusiast, I love exploring new tools and platforms. It's amazing how technology can solve problems and connect people globally."""


# 2a. Words with more than 5 letters
long_words = re.findall(r'\b\w{6,}\b', text)
print("Words with >5 letters:", long_words)

# 2b. Extract numbers
numbers = re.findall(r'\b\d+\b', text)
print("Numbers:", numbers)

# 2c. Capitalized words
capitalized = re.findall(r'\b[A-Z][a-z]+\b', text)
print("Capitalized words:", capitalized)

# 3a. Split into alphabetic words
alpha_words = re.findall(r'\b[a-zA-Z]+\b', text)
print("Alphabetic words:", alpha_words)

# 3b. Words starting with vowel
vowel_words = re.findall(r'\b[aeiouAEIOU]\w+\b', text)
print("Words starting with vowel:", vowel_words)
```

Question 4

```python
import re
def custom_tokenizer(text):
    # Keep contractions and hyphenated words
    tokens = re.findall(r"\b\w+(?:[-']\w+)*\b|\d+\.\d+|\d+", text)
    return tokens

sample_text = "Email me at test@example.com or visit https://example.com. Call +91 9876543210 or 123-456-7890."
tokenized = custom_tokenizer(sample_text)
print("Custom tokens:", tokenized)

# Regex substitutions
cleaned = re.sub(r'\S+@\S+', '<EMAIL>', sample_text)
cleaned = re.sub(r'https?://\S+', '<URL>', cleaned)
cleaned = re.sub(r'\+\d{2}\s\d{10}|(?:\d{3}-){2}\d{4}', '<PHONE>', cleaned)
print("Cleaned text:", cleaned)
```
