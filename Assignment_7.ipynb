Question 1, Part I

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

roll_number = 102317106
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)
```

Part II

```python
print("First 5 rows:")
print(df.head())
print("\nSummary statistics:")
print(df.describe())

print("\nTotal sales per category:")
print(df.sum())
print("\nTotal sales per month:")
print(df.sum(axis=1))

growth = df.diff().mean()
print("\nAverage sales growth between consecutive months:")
print(growth)

df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100

if roll_number % 2 == 0:
    df['Electronics'] = df['Electronics'] * 0.9
else:
    df['Clothing'] = df['Clothing'] * 0.85
```

Part III

```python
plt.figure(figsize=(10, 6))
for category in categories:
    plt.plot(df.index, df[category], label=category)
plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.ylabel('Sales')
plt.show()
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute value
print("Absolute values:")
print(np.abs(array))

# ii. Percentiles
print("\n25th, 50th, 75th percentiles:")
print("Flattened array:", np.percentile(array, [25, 50, 75]))
print("Columns:")
for i in range(array.shape[1]):
    print(f"Column {i}:", np.percentile(array[:, i], [25, 50, 75]))
print("Rows:")
for i in range(array.shape[0]):
    print(f"Row {i}:", np.percentile(array[i, :], [25, 50, 75]))

# iii. Statistics
print("\nStatistics:")
print("Flattened array - Mean:", np.mean(array), "Median:", np.median(array), "Std:", np.std(array))
print("Columns:")
for i in range(array.shape[1]):
    col = array[:, i]
    print(f"Column {i} - Mean:", np.mean(col), "Median:", np.median(col), "Std:", np.std(col))
print("Rows:")
for i in range(array.shape[0]):
    row = array[i, :]
    print(f"Row {i} - Mean:", np.mean(row), "Median:", np.median(row), "Std:", np.std(row))
```

Question 4

```python
def swap_list_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

# Example usage:
my_list = [1, 2, 3, 4, 5]
print("Before swap:", my_list)
swap_list_elements(my_list, 1, 3)
print("After swap:", my_list)
```

Question 5

```python
def swap_set_elements(s, elem1, elem2):
    if elem1 in s and elem2 in s:
        lst = list(s)
        idx1, idx2 = lst.index(elem1), lst.index(elem2)
        lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
        return set(lst)
    return s

my_set = {1, 2, 3, 4, 5}
print("Before swap:", my_set)
my_set = swap_set_elements(my_set, 2, 4)
print("After swap:", my_set)
```
